```python
def get_choices(row):
    choices = {'First Choice':'', 'Second Choice': '', 'Third Choice':''}
    for col in df.columns[5:35]:
        if pd.notna(row[col]):
            choices[row[col]] = col
    return [choices[key] for key in list(choices.keys())]   
```

```python
df['Choices'] = df.apply(get_choices, axis=1)
```
