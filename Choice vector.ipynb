```python
# import necessary libraries

import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
pd.set_option('display.max_rows', None)
```

```python
# save the data in a dataframe

df = pd.read_csv('Test data - Students.csv')
```

```python
# rename topic columns by just the names of the topic, i.e. simplify column names' by removing the question

for i in range(5,35):
    df.rename(columns = {df.columns[i]: df.columns[i][80:-1]},inplace=True)
```

```python
# see the first 5 rows of the dataframe

df.head()
```

```python
# get a list of subjects

subjects = [col for col in df.columns[5:35]]
```

```python
# create the choice_vectors dictionary where keys are Students and the values are their ordered choices

choice_vectors = {}
for index,row in df.iterrows():
    choices = [1,2,3]
    for subject in subjects:
        if row[subject] == 'First Choice':
            choices[0] = subject
        if row[subject] == 'Second Choice':
            choices[1] = subject
        if row[subject] == 'Third Choice':
            choices[2] = subject
    choice_vectors[row['Name']] = choices
  
```
