```python
# Read data

data_mentor = pd.read_csv('unrefined/data-mentors.csv', encoding='utf-8')
data_student = pd.read_csv('unrefined/data-students.csv', encoding='utf-8')
data_pairing = pd.read_csv('unrefined/data-pairings.csv', encoding = 'utf-8')
```

Required changes in the data files:


Both in data_student and data_mentor; Simplify the column names to only subject names by removing the questions These are the 6th-36th indexed columns in data_student, 5-35 in data_mentor



In data_mentor; quantify mentor's position. Master:1, PhD Student:2, Postdoctoral Researcher:3

```python
# 1. Remove the questions in the column names in data_student and data_mentor

for i in range(5,35):
    data_student.rename(columns = {data_student.columns[i+1]: data_student.columns[i+1][80:-1]},inplace=True)
    data_mentor.rename(columns = {data_mentor.columns[i]: data_mentor.columns[i][66:-1]},inplace=True)
```

```python
# 2. Introduce a new column 'Position_numerical' that quantifies mentor's positions

def quantify_position(pos):
    if pos == 'Master':
        return 1
    if pos == 'PhD student':
        return 2
    if pos == 'Postdoctoral researcher':
        return 3

data_mentor.insert(loc = 4, column = 'Position_numeric', value = data_mentor['Position'].apply(quantify_position))
```
