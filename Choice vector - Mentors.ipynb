```python
df = pd.read_csv('df - Mentors.csv')
```

```python
# see the first 5 rows of the dataframe

df.head()
```

```python
# get a list of subjects

subjects = [col for col in df.columns[4:38]]
```

```python
# create the choice_vectors dictionary where keys are Students and the values are their ordered choices

choice_vectors = {}
for index,row in df.iterrows():
    choices = [1,2,3,4]
    for subject in subjects:
        if row[subject] == 'First Choice':
            choices[0] = subject
        if row[subject] == 'Second Choice':
            choices[1] = subject
        if row[subject] == 'Third Choice':
            choices[2] = subject
        if row[subject] == 'Fourth Choice':
            choices[3] = subject
    choice_vectors[row['Name']] = choices
```
